```python
from serpapi import GoogleSearch

def google_search_serpapi(query):

    params = {
        "q": query,
        "hl": "en",
        "gl": "us",
        "api_key": "YOUR_SERPAPI_KEY"  # Replace with your actual API key
    }
    search = GoogleSearch(params)
    results = search.get_dict()
    return [res["link"] for res in results.get("organic_results", [])]


intended_search = input("Enter your search query: ")
results = google_search_serpapi(intended_search)

print("\nSearch Results:")
for idx, result in enumerate(results, 1):
    print(f"{idx}: {result}")
```
